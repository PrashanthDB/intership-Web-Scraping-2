```python
type_quote=[]
start=0
end=10
for page in range(start,end):
    quote_type=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in quote_type:
           type_quote.append(i.text)
next_btn=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]//a") 
next_btn.click()
type_quote
```
